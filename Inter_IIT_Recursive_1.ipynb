```python
max_step=100

directions=[[-1,0,0],[1,0,0],[0,-1,0],[0,1,0],[-1,-1,0],[1,-1,0],[-1,1,0],[1,1,0]]
def new_cmp_lt(self,a,b):
    return a[1]<b[1]
def check_nofly(obj):
  if len(no_fly_regions)==0:
    return False
  for zone in no_fly_regions:
    if zone.is_inside(obj)==True:
      return True
  return False
class vertex:
  def __init__(self,xx=0.0,yy=0.0,zz=0.0):
    self.cordinate=np.array([xx,yy,zz])

  def __lt__(self,other):
    return self.cordinate[2]<other.cordinate[2]
  
  def tupple(self):
    return (self.cordinate[0],self.cordinate[1],self.cordinate[2])

  def euc_distance(self,another_point):
    return np.sqrt(np.sum(np.square(np.subtract(self.cordinate,another_point.cordinate))))

  def is_valid(self):
    return self.cordinate[2]>0

  def min_time(self,point,velocities):
    vx=velocities[0]
    vy=velocities[1]
    vz=velocities[2]
    cost={}
    g={}
    d={}
    max_h=-1e9
    h=0
    init_state=vertex(self.cordinate[0],self.cordinate[1],point.cordinate[2])
    h=max(0,point.cordinate[2]-self.cordinate[2])
    cost[init_state]=0
    g[init_state]=0
    d[init_state]=0
    closed_list=set()
    pq = []
    heapq.heappush(pq, (0, init_state))
    ff=True
    while len(pq)>0 and ff:
      dis=pq[0][0]
      obj=pq[0][1]
      heapq.heappop(pq)
      if obj.tupple() in closed_list:
        continue
      # print(obj.cordinate)
      # max_h=max(max_h,obj.cordinate[2])
      closed_list.add(obj.tupple())
      if np.array_equal(obj.cordinate,point.cordinate):
        cost[obj]=dis
        ff=False
        break
      for ind in directions:
        step=0
        if obj.euc_distance(point)>=10*max_step:
          step=max_step
        elif obj.euc_distance(point)>=10:
          step=int(obj.euc_distance(point)/10)
        else:
          step=1
        flag=True
        # while flag and step>0:
        new_obj=vertex(obj.cordinate[0]+ind[0]*step,obj.cordinate[1]+ind[1]*step,obj.cordinate[2]+ind[2]*step)
          
        if new_obj.tupple() not in closed_list and new_obj.is_valid() and not check_nofly(new_obj): 
          temp=abs(ind[0])+abs(ind[1])+abs(ind[2])
          nstep=0
          if temp==1:
            nstep=1
            if (abs(ind[0])==1 or abs(ind[1])==1) :
              temp=(1*step)/vx
            else:
              temp=(1*step)/vz

          elif temp==2:
            nstep=math.sqrt(2)
            if abs(ind[0])==1 and abs(ind[1])==1:
              temp=(math.sqrt(2)*step)/vx
            else:
              temp=(math.sqrt(2)*step)/min(vx,vz)
          else:
            nstep=math.sqrt(3)
            temp=(math.sqrt(3)*step)/min(vx,vz)

          gnew=g[obj]+temp
          dnew=d[obj]+nstep*step
          hnew=new_obj.euc_distance(point)/min(vx,vz)
          fnew=gnew+hnew
          if new_obj not in cost or cost[new_obj]>fnew:
            cost[new_obj]=fnew
            g[new_obj]=gnew
            d[new_obj]=dnew
            heapq.heappush(pq,(fnew,new_obj))      
      
    return g[obj],d[obj],h   # time,distance,height



    
  
```

```python
class cuboid:
    def __init__(self,a1,a2,a3,a4,a5,a6,a7,a8):
        self.v1=a1
        self.v2=a2 
        self.v3=a3
        self.v4=a4
        self.v5=a5
        self.v6=a6
        self.v7=a7
        self.v8=a8
    def is_inside(self,point):
        u = np.subtract(self.v1.cordinate, self.v2.cordinate)
        v = np.subtract(self.v1.cordinate, self.v4.cordinate)
        w = np.subtract(self.v1.cordinate, self.v5.cordinate)
        if min(np.dot(u, self.v1.cordinate), np.dot(u, self.v2.cordinate)) < np.dot(u, point.cordinate) < max(np.dot(u, self.v1.cordinate), np.dot(u, self.v2.cordinate))\
        and min(np.dot(v, self.v1.cordinate), np.dot(v, self.v4.cordinate)) < np.dot(v, point.cordinate) < max(np.dot(v, self.v1.cordinate), np.dot(v, self.v4.cordinate))\
        and min(np.dot(w, self.v1.cordinate), np.dot(w, self.v5.cordinate)) < np.dot(w,point.cordinate) < max(np.dot(w, self.v1.cordinate), np.dot(w, self.v5.cordinate)):
            return True
        else :
            return False
```

```python
class Item:
    def __init__(self, idd, weight, l, b, h):
        self.id=idd
        self.weight=weight
        self.length = l
        self.breadth = b
        self.height = h
        self.volume = l*b*h
```

```python
class Warehouse:
    def __init__(self,idd,xx,yy,zz,amp):
        self.id=idd
        self.location=vertex(xx,yy,zz)
        self.items=[]
        self.amp=amp
```

```python
class City:
    def __init__(self,idd,xx,yy,zz,item,starttime,endtime,warehouse,day):
        self.id=idd
        self.item =item
        self.day = day
        self.starttime = starttime
        self.warehouse = warehouse
        self.endtime = endtime
        self.location = vertex(xx,yy,zz)
    def __lt__(self, other):
        return self.endtime < other.endtime
```

```python
class Charging_station:
    def __init__(self,idd,xx,yy,zz,max_slots,amp,time_dict):
        self.id=idd
        self.location=vertex(xx,yy,zz)
        self.max_slots=max_slots
        self.amp=amp
        self.time_dict = time_dict
```

```python
!python -m pip install gurobipy
from gurobipy import *
import numpy as np
def coord_No_Fly_Zone (n, sz, P, Q, f, M , MM, bounds, start, end ):
    alpha=[i/sz for i in range(sz+1)]
    A = [i for i in range(2*n+2)]
    L = [i for i in A[1:2*n+2]]
    Zone=[i for i in range(n)]
    C = [(i,j,k) for i in Zone for j in range(2*n+2) for k in range(6)]
    CExtra=[i for i in L]
    OBJs=[i for i in range(2*n+1)]
    mdl = Model('InterIIT')
    x = mdl.addVars(A, lb=-20000,vtype=GRB.CONTINUOUS,name="X")
    y = mdl.addVars(A, lb=-20000,vtype=GRB.CONTINUOUS,name="Y")
    z = mdl.addVars(A,vtype=GRB.CONTINUOUS,name="Z")
    c = mdl.addVars(C, vtype=GRB.BINARY,name="C")
    cextra = mdl.addVars(CExtra, vtype=GRB.BINARY,name="C_Extra")
    obj = mdl.addVars(L,vtype=GRB.CONTINUOUS,name="Obj")

    mdl.modelSense = GRB.MINIMIZE
    mdl.setObjective(quicksum(obj[k] for k in L))

    mdl.addConstr(x[0]==start[0])
    mdl.addConstr(y[0]==start[1])
    mdl.addConstr(z[0]==start[2])
    mdl.addConstr(x[2*n+1]==end[0])
    mdl.addConstr(y[2*n+1]==end[1])
    mdl.addConstr(z[2*n+1]==end[2])

    for i in range(len(alpha)):
        for j in range(len(bounds)):
            for k in L:
                mdl.addConstr(alpha[i]*x[k]+(1-alpha[i])*x[k-1]<=bounds[j][0]+(1-c[j,k,0])*MM )
                mdl.addConstr(alpha[i]*x[k]+(1-alpha[i])*x[k-1]>=bounds[j][1]-(1-c[j,k,1])*MM )
                mdl.addConstr(alpha[i]*y[k]+(1-alpha[i])*y[k-1]<=bounds[j][2]+(1-c[j,k,2])*MM )
                mdl.addConstr(alpha[i]*y[k]+(1-alpha[i])*y[k-1]>=bounds[j][3]-(1-c[j,k,3])*MM )
                mdl.addConstr(alpha[i]*z[k]+(1-alpha[i])*z[k-1]<=bounds[j][4]+(1-c[j,k,4])*MM )
                mdl.addConstr(alpha[i]*z[k]+(1-alpha[i])*z[k-1]>=bounds[j][5]-(1-c[j,k,5])*MM )
    for j in range(len(bounds)):
        for k in range(2*n+2):
            mdl.addConstr(quicksum(c[j,k,l] for l in range(6))>=1)
    for k in L:
        mdl.addConstr( ((x[k]-x[k-1])**2 + (y[k]-y[k-1])**2)/(M-P*f)**2 <= obj[k]**2 )
        mdl.addConstr( ((z[k] - z[k-1])/(M - Q * f)*cextra[k]+(z[k] - z[k-1])/(M + Q * f) *(1-cextra[k])) <= obj[k])
    for k in L:
        mdl.addConstr( z[k] - z[k-1] <= 0 + cextra[k] * MM )
        mdl.addConstr( z[k] - z[k-1] >= 0 - ( 1- cextra[k] ) * MM )

    mdl.params.NonConvex = 2
    mdl.Params.MIPGap = 0.1
    mdl.Params.TimeLimit = 30  # seconds
    mdl.optimize()
    
    XCoord =[]
    YCoord = []
    ZCoord = []
    
    Index = 0;
    XIndexRange = [ 1, 2*n ]
    YIndexRange = [ 2*n+3, 4*n+2 ]
    ZIndexRange = [ 4*n+5, 6*n+4 ]
    for v in mdl.getVars():
        if Index >= XIndexRange[0] and Index <= XIndexRange[1]:
            XCoord.append( v.x )
        if Index >= YIndexRange[0] and Index <= YIndexRange[1]:
            YCoord.append( v.x )
        if Index >= ZIndexRange[0] and Index <= ZIndexRange[1]:
            ZCoord.append( v.x )
        
        Index= Index+1
    
    
    return XCoord , YCoord, ZCoord
```

```python
def go_to_city(Location, curr_status, curr_time, city_index, curr_cost):
    path.append(("City",Location.index))
    temp_1 = copy.deepcopy(curr_status)
    temp_2 = copy.deepcopy(Location)
    temp_3 = curr_time
    temp_4 = curr_cost
    temp_5 = city_index
    x = possible(Location.location, cities[city_index].location, curr_status)
    if x[1] <= curr_status.curr_charge and cities[city_index].item.id in curr_status.payload:
        curr_status.curr_charge -= x[1]
        Location.location = cities[city_index].location
        Location.loc_type = 2
        Location.index = city_index
        curr_time += (x[0] + 180)
        curr_cost += (curr_status.drone.var_cost/(60*60))*x[0]
        if curr_time <= cities[city_index].endtime:
            recur(Location, curr_status, curr_time, city_index+1, curr_cost)
    curr_status = copy.deepcopy(temp_1)
    Location = copy.deepcopy(temp_2)
    curr_time = temp_3
    curr_cost = temp_4
    city_index = temp_5
    path.pop()
```

```python
def freewhen(charging_station, curr_status, curr_time):
    cnt = 0
    starting_time=0
    need = calculate_time_charge(curr_status, charging_station.amp)
    # add charge to
    for i in range(int(curr_time),max_time):
        if charging_station.time_dict[i]<charging_station.max_slots:
            if cnt==0:
                starting_time=i
            cnt+=1
        else:
            if cnt>=need:
                break
            cnt=0
    if cnt>=need:
        return starting_time
    else:
        return -1;
```

```python
def charge_needed(curr_status):

    # return True
    if curr_status.curr_charge/curr_status.drone.battery_capacity<0.75:
        return True
    return False
```

```python
def calculate_time_charge(curr_status,amp):
    # add %
    return int((curr_status.drone.battery_capacity-curr_status.curr_charge)/amp)*(3.6)
```

```python
def freewhen(charging_station, curr_status, curr_time):
    cnt = 0
    starting_time=0
    need = calculate_time_charge(curr_status, charging_station.amp)
    # add charge to
    for i in range(int(curr_time),max_time):
        if charging_station.time_dict[i]<charging_station.max_slots:
            if cnt==0:
                starting_time=i
            cnt+=1
        else:
            if cnt>=need:
                break
            cnt=0
    if cnt>=need:
        return starting_time
    else:
        return -1;
```
